#### **1. Basic Plotting**

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
x=np.linspace(-10,10)
y=x**2-4*x+4
plt.plot(x,y,label=r"$f(x) = x^2 - 4x + 4$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(loc="upper right")
plt.show()
```

#### **2. Sine and Cosine Plot**

```python
x2=np.linspace(0,np.pi)
y1=np.sin(x2)
y2=np.cos(x2)
plt.plot(x2,y1,'g--o',label="$ \\sin(x) $")
plt.plot(x2,y2,'k-.',label="$ \\cos(x) $")
plt.legend()
plt.show()
```

#### **3. Subplots**

```python
x3=np.linspace(0,10)
fig, axes = plt.subplots(2, 2)  # 2x2 grid of subplots
axes[0, 0].plot(x3,x3**3,"r:", label=r"$ f(x) = x^3 $")  # Top-left subplot
axes[0, 0].legend()
axes[0, 1].plot(x3,np.sin(x3), 'g-.', label=r"$ f(x) = \sin(x) $") # Top-right subplot
axes[0, 1].legend()
axes[1, 0].plot(x3,np.e**x3, 'k--', label=r"$ f(x) = e^x $") # Bottom-left subplot
axes[1, 0].legend()
axes[1, 1].plot(x3,np.where(x3!=0, np.log(x3+1),0), label=r"$ f(x) = \log(x+1) $ (for $ x \geq 0 $)") # Bottom-right subplot
axes[1, 1].legend()
plt.show()
```

#### **4. Scatter Plot**

```python
# make a list of random numbers
x4 = np.random.uniform(0, 10, 100)
y4 = np.random.uniform(0, 10, 100)
# Make a list of random markers
markers = ['o', '*', 'v', 'D',  '^', 's','<', '>', 'p', 'h']
marker_choices = np.random.choice(markers, size=100)
plt.figure(figsize=(8, 6))
# Scatter the points
for i in range(100):
    plt.scatter(x4[i], y4[i], marker=marker_choices[i])
# Add Title and axis labels
plt.title("Scatter Plot of 100 Random Points")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
# Add grid
plt.grid(True, linestyle="--", alpha=0.7)
# Show the plot
plt.show()
```

#### **5. Histogram**

```python
values = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(values, bins=30, color="green", edgecolor="red", alpha=0.6)
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

#### **6. 3D Plotting**

```python
x6 = np.linspace(-5, 5, 1000)
y6 = np.linspace(-5, 5, 1000)
xx, yy = np.meshgrid(x6, y6)
f6 = np.cos(xx**2+yy**2)
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot_surface(xx, yy, f6, cmap="cool")
plt.title(r"$ f(x, y) = \cos(x^2 + y^2) $")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

#### **7. Bar Chart**

```python
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
values = [200, 150, 250, 175, 225]
bar_colors = ["yellow", "red", "green","orange",  "grey"]
plt.bar(products, values, color=bar_colors, width=0.6, edgecolor="black")
plt.ylabel("Sales Values")
plt.xlabel("Products")
plt.title("Sales Data")
plt.show()
```

#### **8. Stacked Bar Chart**

```python
values_a = [4, 2, 1, 3]
values_b = [7, 3, 5, 8]
values_c = [3, 6, 4, 5]
time_periods = ["T1", "T2", "T3", "T4"]
plt.bar(time_periods, values_a, label="Category A", color="red")
plt.bar(time_periods, values_b, bottom=values_a, label="Category B", color="green")
plt.bar(time_periods, values_c, bottom=np.array(values_a)+np.array(values_b), label="Category C", color="blue")
plt.legend()
plt.xlabel("Time Periods")
plt.ylabel("Values")
plt.title("Stacked Bar Chart of Categories over Time")
plt.show()
```
